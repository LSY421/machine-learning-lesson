```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
data = pd.read_csv("../data/xigua3.0.csv")
data
```

```python
data.columns
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
data2 = encoder.fit_transform(data[['编号', '色泽', '根蒂', '敲声', '纹理', '脐部', '触感', '好瓜']])
data2
```

```python
data3 = pd.concat([pd.DataFrame(data2, columns=['编号', '色泽', '根蒂', '敲声', '纹理', '脐部', '触感', '好瓜']), data[['密度', '含糖率']]], axis=1)
data3
```

```python
from sklearn import tree
```

```python
model  = tree.DecisionTreeClassifier()
model.fit(data3[['密度','含糖率']], data3['好瓜'])
model.score(data3[['密度','含糖率']], data3['好瓜'])
```

```python
plt.rcParams['figure.dpi']=400
plt.rcParams['font.sans-serif']=['simhei']
tree.plot_tree(model,
               feature_names=['密度','含糖率'],
               class_names=['好瓜','坏瓜'],
               rounded=True,
               filled=True)
plt.show()
```

```python
model  = tree.DecisionTreeClassifier(criterion='entropy')
model.fit(data3[['密度','含糖率']], data3['好瓜'])
model.score(data3[['密度','含糖率']], data3['好瓜'])
plt.rcParams['figure.dpi']=400
plt.rcParams['font.sans-serif']=['simhei']
tree.plot_tree(model,
               feature_names=['密度','含糖率'],
               class_names=['好瓜','坏瓜'],
               rounded=True,
               filled=True)
plt.show()
```

```python
import sklearn.datasets
from sklearn.datasets import load_boston
boston = sklearn.datasets.load_boston()
boston.data
```

```python
boston_pd = pd.DataFrame(boston.data, columns=boston.feature_names)
boston_pd
```

```python
boston_tar_pd = pd.DataFrame(boston.target, columns=['a'])
boston_tar_median = boston_tar_pd.median()[0]
boston_tar_pd['b']=1
boston_tar_pd['b'][boston_tar_pd['a']<boston_tar_median]=0
boston_tar_pd
```

```python
boston_x = boston_pd[['RM','LSTAT']]
boston_y = boston_tar_pd['b']
boston_y
```

```python
from sklearn.model_selection import cross_val_score
model = tree.DecisionTreeClassifier(max_depth=3)
cross_val_score(model, boston_x, boston_y).mean()
```

```python
from sklearn.linear_model import LogisticRegression
cross_val_score(LogisticRegression(), boston_x, boston_y).mean()
```

```python
model  = tree.DecisionTreeClassifier(max_depth=3)
model.fit(boston_x, boston_y)
plt.rcParams['figure.dpi']=400
plt.rcParams['font.sans-serif']=['simhei']
tree.plot_tree(model,
               feature_names=['RM','LSTAT'],
               class_names=['高','低'],
               rounded=True,
               filled=True)
plt.show()
```

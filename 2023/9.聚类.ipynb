```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn
```

```python
from sklearn import cluster
cluster.__all__
```

```python
from sklearn import mixture
mixture.__all__
```

```python
scale=200
x = np.linspace(0, 1, scale + 1)
y = np.linspace(0, 1, scale + 1)
x, y = np.meshgrid(x, y)
disp_input = np.vstack((x.flatten(), y.flatten())).T
```

```python
from sklearn.datasets import make_blobs
x_3, y_3 = make_blobs(n_samples=2000,
                      n_features=2,
                      centers=4,
                      cluster_std=3)

x_3 = (x_3 - x_3.min(axis=0)) / (x_3.max(axis=0) - x_3.min(axis=0))
xb, yb = x_3, y_3
plt.scatter(x_3[:, 0], x_3[:, 1], c=y_3)
plt.show()
```

```python
from sklearn.datasets import make_circles
x_1, y_1 = make_circles(n_samples=1000,
                        noise=0.2,
                        factor=0)
x_1 = (1 + x_1 / x_1.max(axis=0)) / 2
xb, yb = x_1, y_1
plt.scatter(x_1[:, 0], x_1[:, 1], c=y_1)
plt.show()
```

```python
from sklearn.datasets import make_swiss_roll

x_4, y_4 = make_swiss_roll(n_samples=500,
                           noise=0.2,
                           random_state=2)

x_4 = np.array([x_4[:,0],x_4[:,2], y_4]).T
x_4[:,2] = 1

x_4 = np.row_stack((x_4, -x_4))

y_4 = x_4[:,2]
y_4 = (y_4+abs(y_4))/2

x_4 = x_4[:,:2]

x_4 = (x_4 - x_4.min(axis=0)) / (x_4.max(axis=0) - x_4.min(axis=0))
xb,yb=x_4,y_4
plt.scatter(x_4[:, 0], x_4[:, 1], c=y_4, s=10)
plt.show()
```

```python
model = cluster.KMeans(n_clusters=2)
yb_predict = model.fit_predict(xb)
disp_predict = model.predict(disp_input)
plt.scatter(disp_input[:, 0], disp_input[:, 1], c=disp_predict, cmap='gray', alpha=0.3, s=1)
plt.scatter(xb[:, 0], xb[:, 1], c=yb_predict)
plt.show()
```

```python
# scale=200
# x = np.linspace(-100, 100, scale + 1)
# y = np.linspace(-100, 100, scale + 1)
# x, y = np.meshgrid(x, y)
# disp_input = np.vstack((x.flatten(), y.flatten())).T

model = mixture.GaussianMixture(n_components=2)
yb_predict = model.fit_predict(xb)
disp_predict = model.predict(disp_input)
plt.scatter(disp_input[:, 0], disp_input[:, 1], c=disp_predict, cmap='gray', alpha=0.3, s=1)
plt.scatter(xb[:, 0], xb[:, 1], c=yb_predict)
plt.show()
```

```python
plt.rcParams['figure.dpi']=200
model = cluster.DBSCAN(eps=0.03,
                       min_samples=8)
yb_predict = model.fit_predict(xb)
plt.scatter(xb[:, 0], xb[:, 1], c=yb_predict)
plt.show()
```

```python
plt.rcParams['figure.dpi']=200
model = cluster.AgglomerativeClustering(n_clusters=2)
yb_predict = model.fit_predict(xb)
plt.scatter(xb[:, 0], xb[:, 1], c=yb_predict)
plt.show()
```

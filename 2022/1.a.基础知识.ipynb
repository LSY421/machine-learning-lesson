```python
xigua = pd.read_csv("../data/xigua3.0.csv", index_col='编号')
xigua
```

```python
xigua.to_excel("../data/xigua3.0.xlsx")
xigua.to_json("../data/xigua3.0.json")
xigua.to_pickle("../data/xigua3.0.pkl")
print("done")
```

```python
xigua_2_1 = pd.get_dummies(xigua[['敲声', '纹理', '脐部', '触感', '色泽','根蒂']])
```

```python
xigua_2_2 = xigua['好瓜'].replace({'是':1, '否':0})
```

```python
xigua.columns
```

```python
xigua_2 = pd.concat([xigua[['密度', '含糖率']], xigua_2_1, xigua_2_2], axis=1)
xigua_2
```

```python
import matplotlib.pyplot as plt
plt.rcdefaults()
plt.rcParams['font.family']=['SimHei']
plt.rcParams['font.size']=5
plt.rcParams['figure.dpi']=200
xigua.hist()
plt.show()
```

```python
xigua_2[['密度','含糖率']].plot()
```

```python
xigua_2[['密度','含糖率']].plot(kind='bar')
```

```python
xigua_2[['密度','含糖率','好瓜']].plot(kind='scatter', x='密度',y='含糖率', c='好瓜', cmap='cool')
plt.show()
```

```python
plt.scatter(x=xigua_2['密度'],y=xigua_2['含糖率'], c=xigua_2['好瓜'], cmap='cool')
for idx, row in xigua_2.iterrows():
    plt.text(row['密度']+0.005,row['含糖率']-0.002,str(idx))
plt.show()
```

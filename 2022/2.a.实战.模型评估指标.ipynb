```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
```

```python
a = np.random.randint(0, 2, 10)
b = np.ones(10)
(a-b).sum()
```

```python
result = []
result.append(['acc', accuracy_score(a,b)])
result.append(['recall', recall_score(a,b)])
result.append(['f1', f1_score(a,b)])
result.append(['roc_auc', roc_auc_score(a,b)])
result_pd  = pd.DataFrame(result)
result_pd
```

```python
plt.rcdefaults()
plt.rcParams['figure.dpi']=200
plt.bar(result_pd[0],result_pd[1], color='lime',alpha=0.6)
plt.show()
```

```python
a = np.random.randint(0, 2, 10)
b = np.ones(10)
result = []
result.append(['acc', accuracy_score(a,b)])
result.append(['recall', recall_score(a,b)])
result.append(['f1', f1_score(a,b)])
result.append(['roc_auc', roc_auc_score(a,b)])
result_pd  = pd.DataFrame(result)
plt.rcdefaults()
plt.rcParams['figure.dpi']=100
plt.title("err="+str((a-b).sum()))
plt.bar(result_pd[0],result_pd[1], color='lime',alpha=0.6)
plt.show()
```

```python
x = np.linspace(0,1,101)
y1 = x*x
y2 = x*x + np.random.rand(101)/10
y1, y2
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_log_error
from sklearn.metrics import median_absolute_error
from sklearn.metrics import r2_score
```

```python
for scale in range(1,11):
    x = np.linspace(0,1,101)
    y1 = x*x
    y2 = x*x + np.random.rand(101)/scale
    result = []
    result.append(['MAE', mean_absolute_error(y1, y2)])
    result.append(['MSE', mean_squared_error(y1, y2)])
    result.append(['RMSE', np.sqrt(mean_squared_error(y1, y2))])
    result.append(['MSLE', mean_squared_log_error(y1, y2)])
    result.append(['MedAE', median_absolute_error(y1, y2)])
    result.append(['R2', r2_score(y1, y2)])
    result_pd = pd.DataFrame(result)
    plt.rcdefaults()
    plt.rcParams['figure.dpi']=100
    plt.title("scale="+str(scale))
    plt.bar(result_pd[0],result_pd[1], color='lime',alpha=0.6)
    plt.show()
```

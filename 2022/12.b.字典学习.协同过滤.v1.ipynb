```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn
import sklearn.datasets
import sklearn.decomposition as de
de.__all__
```

```python
ratings_1 = pd.read_csv("../data/ml-1m/ratings.dat",
                        sep="::",
                        header=None,
                        engine='python')
ratings_1.columns = ['UserID','MovieID','Rating','Timestamp']
ratings_1
```

```python
ratings_2 = ratings_1[ratings_1['Rating']>3]
ratings_2
```

```python
ratings_2_onehot = pd.get_dummies(ratings_2['MovieID'])
ratings_2_onehot
```

```python
ratings_3 = pd.concat([ratings_2['UserID'], ratings_2_onehot],axis=1)
ratings_3
```

```python
ratings_4 = ratings_3.groupby(by=['UserID']).sum()
ratings_4
```

```python
ratings_4 = pd.read_csv("../data/ml-1m/ratings_4.csv", index_col='UserID')
ratings_4
```

```python
ratings_4.loc[1][ratings_4.loc[1]>0]
```

```python
model = de.TruncatedSVD(n_components=100)
ratings_4_svd = model.fit_transform(ratings_4)
ratings_4_svd
```

```python
ratings_4_svd_inverse = model.inverse_transform(ratings_4_svd)
ratings_4_svd_inverse
```

```python
ratings_4_svd_inverse_pd = pd.DataFrame(ratings_4_svd_inverse, columns=ratings_4.columns)
ratings_4_svd_inverse_pd.index = ratings_4.index
ratings_4_svd_inverse_pd
```

```python
user_prefs = pd.DataFrame()
user_prefs.index = ratings_4.index
for (idx, row),i in zip(ratings_4_svd_inverse_pd.iterrows(),range(100000000000)):
    # print([x for x in row.argsort()[-30:].values[::-1]])
    user_prefs.loc[idx, 'prefs'] = ",".join([str(x) for x in row.argsort()[-30:].values[::-1]])
user_prefs
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn
import sklearn.datasets
import sklearn.decomposition as de
de.__all__
```

```python
ratings_1 = pd.read_csv("../data/ml-1m/ratings.dat",
                        sep="::",
                        header=None,
                        engine='python')
ratings_1.columns = ['UserID','MovieID','Rating','Timestamp']
ratings_1
```

```python
ratings_2_onehot = pd.get_dummies(ratings_1['MovieID'])
```

```python
ratings_2_onehot
```

```python
ratings_2_onehot.loc[0,1194]
```

```python
ratings_2 = ratings_2_onehot.multiply(ratings_1['Rating'], axis='index')
ratings_2
```

```python
ratings_2.loc[0,1193]
```

```python
ratings_3 = pd.concat([ratings_2, ratings_1['UserID']], axis=1)
ratings_3 = ratings_3.groupby(by='UserID').sum()
ratings_3
```

```python
ratings_4 = ratings_3.replace({3:0,2:0,1:0})
ratings_4
```

```python
ratings_4 = ratings_3.replace({5:2, 4:1, 3:0, 2:-1, 1:-2})
ratings_4
```

```python
ratings_4.loc[1,513]
```

```python
model = sklearn.decomposition.TruncatedSVD(n_components=100)
ratings_4_svd = model.fit_transform(ratings_4)
ratings_4_svd
```

```python
user_prefs = model.inverse_transform(ratings_4_svd)
user_prefs_pd = pd.DataFrame(user_prefs, columns=ratings_4.columns)
user_prefs_pd.index = ratings_4.index
user_prefs_pd
```

```python
user_history = pd.DataFrame()
user_history.index = user_prefs_pd.index
for (idx,row), i in zip(ratings_3.iterrows(), range(10000000000000)):
    user_history.loc[idx, 'history'] = row[row>0].index.values
user_history
```

```python
# 用户的电影偏好性
user_prefs_movieid = pd.DataFrame()
user_prefs_movieid.index = user_prefs_pd.index
for (idx,row),(idx2, row2), i in zip(user_prefs_pd.iterrows(),user_history.iterrows(), range(1000000000)):
    # print(row.argsort()[-30:].values[::-1])
    row_top = [str(x+1) for x in row.argsort()[-30:].values[::-1] if x+1 not in row2['history']]
    user_prefs_movieid.loc[idx, 'pref'] = ",".join(row_top)
user_prefs_movieid
```

```python
user_prefs_movieid.to_csv("../data/ml-1m/user_prefs_2.csv")
```

```python
users_1 = pd.read_csv("../data/ml-1m/users.dat",
                        sep="::",
                        header=None,
                        engine='python')
users_1.columns = ['UserID','Gender','Age','Occupation','Zip-code']
users_1.index = users_1['UserID']
users_1['Gender'] = users_1['Gender'].replace({"F":1,"M":0})
users_1
```

```python
ratings_5 = pd.concat([ratings_4, users_1[['Gender','Age','Occupation']]], axis=1)
ratings_5
```

```python
model = sklearn.decomposition.TruncatedSVD(n_components=100)
ratings_5_svd = model.fit_transform(ratings_5)
user_prefs = model.inverse_transform(ratings_5_svd)
user_prefs_pd = pd.DataFrame(user_prefs, columns=ratings_5.columns)
user_prefs_pd.index = ratings_5.index
user_prefs_pd
```

```python
user_prefs_pd = user_prefs_pd.iloc[:,:-3]
user_prefs_pd
```

```python
# 用户的电影偏好性
user_prefs_movieid = pd.DataFrame()
user_prefs_movieid.index = user_prefs_pd.index
for (idx, row), (idx2, row2), i in zip(user_prefs_pd.iterrows(), user_history.iterrows(), range(1000000000)):
    # print(row.argsort()[-30:].values[::-1])
    row_top = [str(x + 1) for x in row.argsort()[-30:].values[::-1] if x + 1 not in row2['history']]
    user_prefs_movieid.loc[idx, 'pref'] = ",".join(row_top)
user_prefs_movieid.to_csv("../data/ml-1m/user_prefs_3.csv")
user_prefs_movieid
```

```python
model.components_
```

```python
components_pd = pd.DataFrame(model.components_, columns=ratings_5.columns)
components_pd
```

```python
components_corr = components_pd.corr().abs()
components_corr
```

```python
movie_sim_pd = pd.DataFrame()
movie_sim_pd.index = components_corr.index
for (idx, row), i in zip(components_corr.iterrows(), range(1000000000)):
    row_top = [str(x+1) for x in row.argsort()[-6:-1].values[::-1]]
    movie_sim_pd.loc[idx, 'corr'] = ",".join(row_top)
movie_sim_pd.to_csv("../data/ml-1m/movie_sim_pd.csv")
movie_sim_pd
```

```python
movies_1 = pd.read_csv("../data/ml-1m/movies.dat",
                        sep="::",
                        header=None,
                        engine='python',
                       encoding='ISO8859')
movies_1.columns = ['MovieID','Title','Genres']
movies_1.index = movies_1['MovieID']
movies_1
```

```python
for (idx, row), i in zip(movie_sim_pd.iterrows(), range(100000000000000)):
    row_name = [ movies_1.loc[int(id),'Title'] for id in row['corr'].split(",") if int(id) in movies_1.index]
    # print(row_name)
    movie_sim_pd.loc[idx, 'name'] = movies_1.loc[idx,'Title'] if idx in movies_1.index else 'None'
    movie_sim_pd.loc[idx, 'corrnames'] = ";;".join(row_name)
movie_sim_pd.to_csv("../data/ml-1m/movie_sim_name_pd.csv")
movie_sim_pd
```

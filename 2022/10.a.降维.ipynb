```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn
import sklearn.datasets
```

```python
from sklearn.datasets import fetch_california_housing
data = fetch_california_housing()
data.data
```

```python
import sklearn.decomposition
print(",".join(sklearn.decomposition.__all__))
import sklearn.manifold
print(",".join(sklearn.manifold.__all__))
```

```python
plt.rcParams['figure.dpi'] = 200
plt.rcParams['font.sans-serif'] = ['simhei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
from sklearn.decomposition import PCA
model = PCA(n_components=2)
result_pca = model.fit_transform(data.data)
plt.scatter(result_pca[:, 0], result_pca[:, 1], c=data.target, cmap='cool')
plt.show()
```

```python
# 'linear', 'poly', 'rbf', 'sigmoid', 'cosine', 'precomputed'
from sklearn.decomposition import KernelPCA
for kernel in ['linear', 'poly', 'rbf', 'sigmoid', 'cosine']:
    model = KernelPCA(n_components=2, kernel=kernel, n_jobs=-1)
    result_pca = model.fit_transform(data.data)
    plt.scatter(result_pca[:, 0], result_pca[:, 1], c=data.target, cmap='cool')
    plt.title(kernel)
    plt.show()
```

```python
from sklearn.manifold import MDS
model = MDS(n_components=2, n_jobs=-1)
result_pca = model.fit_transform(data.data)
plt.scatter(result_pca[:, 0], result_pca[:, 1], c=data.target, cmap='cool')
plt.show()
```

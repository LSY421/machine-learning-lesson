```python
xy = np.random.rand(10000, 3)
plt.rcParams['figure.dpi'] = 200
plt.scatter(xy[:, 0], xy[:, 1], c=xy[:, 2], s=1)
plt.show()
```

```python
x, y = sklearn.datasets.make_blobs(n_samples=10000,
                                   n_features=2,
                                   centers=10)
plt.scatter(x[:, 0], x[:, 1], c=y, s=1)
plt.show()
```

```python
x, y = sklearn.datasets.make_circles(n_samples=10000,
                                     noise=0.03
                                     )
plt.scatter(x[:, 0], x[:, 1], c=y, s=1)
plt.show()
```

```python
x, y = sklearn.datasets.make_moons(n_samples=10000,
                                     noise=0.03
                                     )
plt.scatter(x[:, 0], x[:, 1], c=y, s=1)
plt.show()
```
